# **Final Project EDA**

```python
import pandas as pd
import numpy as np
import plotly.express as px
```

# Load the Dataset

```python
song_dataframe = pd.read_csv('BestSongsonSpotify.csv', sep=';')
```

```python
song_dataframe.head()
```

# Cleaning the Data

```python
columns_to_drop = ['dB']
song_dataframe.drop(columns_to_drop, axis=1, inplace = True)
song_dataframe.head()
```

```python
song_dataframe.dropna(inplace=True)
song_dataframe.reset_index(drop=True, inplace=True)
```

# Visualizations

Is pop the most popular genre?

```python
# fig = px.scatter(song_dataframe, x = 'top genre', y = 'popularity', title = 'Popularity of Songs based on Genres')
# fig.show()

# df = song_dataframe.groupby(['top genre']).mean().reset_index()
# fig = px.scatter(df, x = 'top genre', y = 'popularity', title = 'Popularity of Songs based on Genres')
# fig.show()

# df = song_dataframe.groupby(['top genre']).mean().reset_index()
# fig = px.bar(df, x = 'top genre', y = 'popularity')
# fig.show()

df = song_dataframe.groupby(['top genre']).mean().reset_index()
fig = px.bar(df, x= 'top genre', y = 'popularity', color = 'top genre')
fig.show()
```

```python
#pd.DataFrame(song_dataframe[song_dataframe['popularity'] >= 90]['top genre'].value_counts()).plot.pie(subplots=True)

song_dataframe.replace('k-pop girl group', 'k-pop', inplace=True)
fig = px.pie(song_dataframe[song_dataframe['popularity'] >= 90], names='top genre', title='Popularity of top genres')
fig.show()
```

Do more recent songs have higher "liveness"?

```python
df = song_dataframe.groupby(['year']).mean().reset_index()
fig = px.bar(df, x='year', y='liveness', title = 'average liveness of songs by year')
fig.show()
```

Are recent songs more popular?

```python
fig = px.scatter(song_dataframe, x = 'year', y = 'popularity', title = 'Popularity of Songs Based on the Year they were Released')
fig.show()
```

```python
df = song_dataframe.groupby(['year']).mean().reset_index()
fig = px.line(df, x = 'year', y = 'popularity', title = 'Popularity of Songs Based on the Year they were Released')
fig.show()
```

Are songs with higher danceability more popular?

```python
# fig = px.scatter(song_dataframe, x = 'popularity', y = 'danceability ', title = 'popularity vs. danceability')
# fig.show()

fig = px.scatter(song_dataframe, x = 'popularity', y = 'danceability ', title = 'popularity vs. danceability', color = 'top genre')
fig.show()
```

Do songs with a higher BPM have higher energy?

```python
# fig = px.bar(song_dataframe, x='bpm', y='energy')
# fig.show()
df = song_dataframe.groupby(['bpm']).mean().reset_index()
fig = px.bar(df, x='bpm', y='energy', title='energy of songs by bpm')
fig.show()
```

```python
df = song_dataframe.groupby(['bpm']).mean().reset_index()
fig = px.line(df, x='bpm', y='energy', title='energy of songs by bpm')
fig.show()
```

How is the popularity of songs related to its speechiness?

```python
# fig = px.bar(song_dataframe, x='speechiness ', y='popularity', title='popularity of songs by speechiness')
# fig.show()

df = song_dataframe.groupby(['speechiness ']).mean().reset_index()
fig = px.bar(df, x='speechiness ', y='popularity', title='popularity of songs by speechiness')
fig.show()
```

```python
fig = px.scatter(song_dataframe, x='speechiness ', y='popularity', title='popularity of songs by speechiness')
fig.show()
```

How are valence and danceability related?

```python
fig = px.scatter(song_dataframe, x = 'valence', y = 'danceability ', title = 'valence vs. danceability', color = 'top genre')
fig.show()
```
